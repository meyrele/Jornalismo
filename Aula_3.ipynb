**Manipulação de strings**

```python
nome = input('Qual seu nome? ')
nome = nome.capitalize()
print(f'Olá, {nome}! Como vai?')
```

Outra forma:

```python
nome = input('Qual seu nome? ').capitalize()
print(f'Olá, {nome}! Como vai?')
```

```python
nome = input('Qual seu nome? ').title()
print(f'Olá, {nome}! Como vai?')
```

```python
nome = input('Qual seu nome? ').upper()
print(f'Olá, {nome}! Como vai?')
```

```python
nome = input('Qual seu nome? ').lower()
print(f'Olá, {nome}! Como vai?')
```

```python
nome = input('Qual seu nome? ').strip() # tira espaços em branco desnecessários no começo e no final
print(f'Olá, {nome}! Como vai?')
```

```python
nome = input('Qual seu nome? ').title().strip()
print(f'Olá, {nome}! Como vai?')
```

**Abrindo e lendo arquivos**

```python
arquivo = open("GMT20210806-222707_Recording.txt")
for linha in arquivo:
  partes = linha.strip().split("\t") #Lista de strings contendo hora, nome e mensagem
  if len(partes) == 3:
    print(partes[1])
```

```python
import csv
arquivo = open("brasil.csv")
for linha in csv.DictReader(arquivo):
  #print(linha)
  print(f"{linha['municipio']}/{linha['estado']} possui {linha['habitantes']} habitantes.")
```
